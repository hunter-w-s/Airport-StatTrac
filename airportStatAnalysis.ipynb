```python
import pandas as pd
import seaborn as sb
```

```python
deps = df = pd.read_csv('depData.csv', sep=',',converters={'day':str, 'month':str, 'year':str, 'dow':str,'hour':str, 'min':str, 'depICAO':str, 'craftType':str})
deps.drop(columns = df.columns[0], axis = 1, inplace= True)
deps.head()
```

```python
arrs = df = pd.read_csv('arrData.csv', sep=',',converters={'day':str, 'month':str, 'year':str, 'dow':str,'hour':str, 'min':str, 'depICAO':str,  'craftType':str})
arrs.drop(columns = df.columns[0], axis = 1, inplace= True)
arrs.head()
```

```python
counts = deps['dow'].value_counts().rename_axis('Day of Week').reset_index(name='Flights')
ax = sb.barplot(x='Day of Week', y='Flights', data=counts, order=['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday'])
```
